Temos a Ordem Correta

`````python
import moviepy.editor

# Converts into more readable format
def convert(seconds):
    hours = seconds // 3600
    seconds %= 3600
    mins = seconds // 60
    seconds %= 60
    return hours, mins, seconds

def get_convert(hours, mins, secs):
    if(hours == 0):
        return str(mins).zfill(2) + 'm' + str(secs).zfill(2) + 's'
    else:
        return str(hours).zfill(2) + 'h' + str(mins).zfill(2) + 'm' + str(secs).zfill(2) + 's'

def get_duraction_of_file(filepath):
#     print('FILE-PATH ::', filepath)
    # Create an object by passing the location as a string
    video = moviepy.editor.VideoFileClip(filepath)
    # Contains the duration of the video in terms of seconds
    video_duration = int(video.duration)
    hours, mins, secs = convert(video_duration)
#     print("Hours:", hours, "| Minutes:", mins, "| Seconds:", secs)
    return get_convert(hours, mins, secs), video_duration
    
`````

## Iterar em todas as pastas (processo completo)

`````python
# Inicializar estrutura
course_tree = {
    'name_course': root_course,
    'time_course': '',
    'time_course_seconds': 0,
    'number_of_modules': 0,
    'modules' : [],
}
`````

`````python
folders_sorted_of_course = maping_sorted_dict
folders_sorted_of_course
`````

`````python
for number_of_module, filepath_module in folders_sorted_of_course:
    print('\n====================', number_of_module, filepath_module, '\n')
    # initialize module tree
    module_tree = {
        'name': filepath_module,
        'number': number_of_module,
        'time_module': '',
        'time_module_seconds': 0,
        'number_of_videos': 0,
        'files': [],
    }
    
    # Get and  Analyse videos
    video_files = list(filter(lambda x: x.split('.')[-1] in filter_videos_files, list_files(filepath_module)))
    for el in video_files:
        video_name = (os.path.splitext(el)[0]).split('\\')[-1]
        print(video_name)
        video_tree = {
            'name': '',
            'time_video': '',
            'time_video_seconds': '',
        }
        time_string, video_seconds = get_duraction_of_file(el)
        video_tree['time_video_seconds'] = video_seconds
        video_tree['time_video'] = time_string
        video_tree['name'] = video_name
        print('video_tree ::', video_tree, '\n')
        module_tree['files'].append(video_tree)
        module_tree['number_of_videos'] += 1
        module_tree['time_module_seconds'] += video_seconds

    h, m, s = convert(module_tree['time_module_seconds'])
    module_tree['time_module'] = get_convert(h, m, s)
    
    # END
    print('\n => module_tree ::')
    print(json.dumps(module_tree, indent=4))
    course_tree['modules'].append(module_tree)
`````

`````python
# itera para pegar tempo total
for el in course_tree['modules']:
    course_tree['time_course_seconds'] += el['time_module_seconds']
    course_tree['number_of_modules'] += 1

h, m, s = convert(course_tree['time_course_seconds'])
course_tree['time_course'] = get_convert(h, m, s)
`````
