# Snippets

`````python
# Converts into more readable format 
def convert(seconds):
    hours = seconds // 3600
    seconds %= 3600
    mins = seconds // 60
    seconds %= 60
    return hours, mins, seconds

# format: 00h00m00s
def get_convert(hours, mins, secs):
    if(hours == 0):
        return str(mins).zfill(2) + 'm' + str(secs).zfill(2) + 's'
    else:
        return str(hours).zfill(2) + 'h' + str(mins).zfill(2) + 'm' + str(secs).zfill(2) + 's'

# get duraction of video-file
def get_duraction_of_file(filepath):
    video_duration = 6000
    video = moviepy.editor.VideoFileClip(filepath)
    video_duration = int(video.duration)
    hours, mins, secs = convert(video_duration)
    return get_convert(hours, mins, secs), video_duration
`````

`````python
def list_files(dir):
    r = []
    for root, dirs, files in os.walk(dir):
        for name in files:
            r.append(os.path.join(root, name))
    return r
`````

## Sort Folders (Insira código aqui) in Dict Structure

**A função `get_number_of_folder` muda para cada curso: é como pegar o número do módulo no nomes das pastas**

Infelizmente, vai ler na ordem, 1 => 10 => 11 ... , 2 => 20 => 21 ,... ,  e nao queremos isso. ENtâ vamos criar uma lsita de tuplas, onde cada tulpla te tamanho 3 e segue a ordem:

````
(numero, nome-completo, nome-sem-enumeracao)
````

**OBS:** Deverá ser alterado apra cada estrutura de pasta

`````python
# Essas regras vão variar para cada curso
def get_number_of_folder(dir_name):
    ### Se a pasta for tiver o nome ... THE NUMBER
    ## 'Módulo 1 - Introdução'
    ## CÓDIGO: dir_name.split('-')[0].split(' ')[1].zfill(2)
    
    ### Se a pasta for tivre o nome ... THE NUMBER
    ## '06 Detecção e Segmentação de Objetos'
    ## CÓDIGO: dir_name.split(' ')[0].strip().zfill(2)
    
    
    return dir_name.split(' ')[0].strip().zfill(2)

def get_name_of_folder(dir_name):
    ### Se a pasta for tiver o nome ... THE NAME
    ## 'Módulo 1 - Introdução'
    ## CÓDIGO: dir_name.split('-')[1].strip()
    
    ### Se a pasta for tiver o nome ... THE NAME
    ## '06 Detecção e Segmentação de Objetos'
    ## CÓDIGO: " ".join( dir_name.split(' ')[1:] )
    
    
    return " ".join( dir_name.split(' ')[1:] )
`````

`````python
# Tetando a saida dessas funções:
example_folder = '06 Detecção e Segmentação de Objetos'

print('NUMBER:', get_number_of_folder(example_folder))
print('NAME:', get_name_of_folder(example_folder))
`````

`````python
# maping_sorted_dict = deve terminar como uma lsita de tuplas 
#   no seguinte formato:
#   (number_int, (nome_full_pasta, nome_modulo))
maping_sorted_dict = {}
for el in all_dirs:
    number_str = get_number_of_folder(el)
    maping_sorted_dict[int(number_str)] = (el, get_name_of_folder(el))

maping_sorted_dict = sorted(maping_sorted_dict.items(), key=lambda x: x[0])
maping_sorted_dict
`````

# Execute Processe - Generate Folders Structure

Devo iterarra e criar uma super esttutura de ados que tera´todos os dados, e no final, eu faço o write nos arquivos, tanto no Readme quanto no repsctivo course/01-module.md

`````python
# Recuperando todas as pastas do curso
folders_sorted_of_course = maping_sorted_dict
folders_sorted_of_course
`````

`````python
video_example = '1 - Transfer Learning Avan\u00e7ado com Fine-Tuning'

# Ordena o attr 'files' pelo name. Tem que converter para int ou .zfill(2)
def get_number_of_video(str_video):
    # deve retornar a enumeraçao do video com 2 casas
    
    ## SE VIDEO FOR: 
    # '1 - Transfer Learning Avan\u00e7ado com Fine-Tuning'
    ## CÓDIGO: str_video.split('-')[0].strip().zfill(2)
    
    
    return str_video.split('-')[0].strip().zfill(2)
    

def sort_module_tree_files(module_tree):
    module_tree['files'] = sorted(
        module_tree['files'], key=lambda x: get_number_of_video(x['name'])
    )
    return module_tree 

print(get_number_of_video(video_example))
`````

`````python
# Inicializar estrutura do curso inteiro
course_tree = {
    'name_course': root_course,
    'time_course': '',
    'time_course_seconds': 0,
    'number_of_modules': 0,
    'modules' : [],
}

# Files to filter
filter_videos_files = ['mp4','ts', 'MP4']

# Iterar sobre as pastas do curso
# filepath_module com numero, sem numero
for number_of_module, filepath_module in folders_sorted_of_course:
    
    filepath_name, numberless_name = filepath_module[0], filepath_module[1]
    print('\n====================', number_of_module, numberless_name, '\n')
    
    # initialize module tree
    module_tree = {
        'name': filepath_name,
        'number': number_of_module,
        'finalize_name': str(number_of_module).zfill(2) + ' - ' + numberless_name,
        'time_module': '',
        'time_module_seconds': 0,
        'number_of_videos': 0,
        'files': [],
    }
    print('-----------',  str(number_of_module).zfill(2) + ' - ' + numberless_name)
    
    # Get and  Analyse videos
    video_files = list(filter(lambda x: x.split('.')[-1] in filter_videos_files, list_files(filepath_name)))
    for el in video_files:
        video_name = (os.path.splitext(el)[0]).split(separator)[-1]
        print(video_name)
        video_tree = {
            'name': '',
            'time_video': '',
            'time_video_seconds': '',
        }
        time_string, video_seconds = get_duraction_of_file(el)
        # save video data
        video_tree['time_video_seconds'] = video_seconds
        video_tree['time_video'] = time_string
        video_tree['name'] = video_name
        print('video_tree ::', video_tree, '\n')
        # save in module data
        module_tree['files'].append(video_tree)
        module_tree['number_of_videos'] += 1
        module_tree['time_module_seconds'] += video_seconds

    h, m, s = convert(module_tree['time_module_seconds'])
    module_tree['time_module'] = get_convert(h, m, s)
    
    # Ordena os 'files' do 'module_tree'
    #   , pois as vezes não está ordenado
    module_tree = sort_module_tree_files(module_tree)
    
    # END
    print('\n => module_tree ::')
    print(json.dumps(module_tree, indent=4))
    
    course_tree['modules'].append(module_tree)
`````

# Calcula tempo total e salva em `.json`

`````python
# itera para pegar tempo total
for el in course_tree['modules']:
    course_tree['time_course_seconds'] += el['time_module_seconds']
    course_tree['number_of_modules'] += 1

h, m, s = convert(course_tree['time_course_seconds'])
course_tree['time_course'] = get_convert(h, m, s)

# Salvar em JSON
with open('data.json', 'w') as fp:
    json.dump(course_tree, fp)
`````

# Create `README-GLOBAL.md`

`````python
with open("README-GLOBAL.md",'w', encoding = 'utf-8') as f:
    f.write("# {}\n\n".format(course_tree['name_course']))
    f.write("Tempo: {} || Quantidade de Módulos {}\n\n".format(
        course_tree['time_course'], course_tree['number_of_modules'],))
    f.write("## File Tree\n\n")
    for el in course_tree['modules']:
        f.write("├─── {} ({})\n".format(el['finalize_name'], el['time_module']))
        for il in el['files']:
            f.write("│       {} ({})\n".format(il['name'], il['time_video']))
        f.write('│\n')
`````

# Create individual `README.md` to each module

`````python
# Escrevendo N Modulos
for el in course_tree['modules']:
    # str(el['number']).zfill(2) + '-'
    file_module_name = el['finalize_name'] + '.md'
    with open(file_module_name, 'w', encoding = 'utf-8') as f:
        f.write("# {}\n\n".format(el['finalize_name']))
        f.write("Tempo: {} || Quantidade de Vídeos {}\n\n".format(
        el['time_module'], el['number_of_videos'],))
        for il in el['files']:
            f.write("## {} ({})\n\n\n\n".format(il['name'], il['time_video']))  
`````
