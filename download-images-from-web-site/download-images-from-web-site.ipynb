```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os 
```

```python
links = [
    'https://refactoring.guru/pt-br/design-patterns/behavioral-patterns',
    'https://refactoring.guru/pt-br/design-patterns/structural-patterns',
    'https://refactoring.guru/pt-br/design-patterns/creational-patterns'
]
```

```python
base_url = 'https://refactoring.guru/'
img_links = []

for link in links:
    response = requests.get(link)
#     print(response.text)
    soup_page = BeautifulSoup(response.text)
    span_tags = soup_page.find_all('span', attrs = {'class' : 'pattern-image'})
    for l in span_tags:
        img_src = base_url + l.find('img')['src']
        img_links.append(img_src)

print(img_links)
```

```python
img_links[0].split('/')[-1].split('?')[0].replace('-mini','')
```

```python
for img in img_links:
    print(img.split('/')[-1].split('?')[0].replace('-mini',''))
```

```python
for img in img_links:
    r = requests.get(img)
    img_name = img.split('/')[-1].split('?')[0].replace('-mini','')
    with open(img_name, "wb") as f:
        print('GENERATE IMAGE', img_name, 'FROM\n', img,'\n')
        f.write(r.content)
```

# Generate MarkDown

```python
base_url = 'https://refactoring.guru/'
img_links = []

for link in links:
    response = requests.get(link)
#     print(response.text)
    soup_page = BeautifulSoup(response.text)
    span_tags = soup_page.find_all('a', attrs = {'class' : 'pattern-card-extended'})
    for l in span_tags:
        img_desc = l.find('span', class_='pattern-card-bottom').find('p').get_text()
        img_name = l.find('span', class_='pattern-name').get_text()
        img_src = base_url + l.find('img')['src']
        img_links.append( (img_name, img_src, img_desc))



const = '![](https://github.com/rafanthx13/design-patterns-study/blob/main/imgs/all-design-patterns/%%)'

for l in img_links:
    print( '**' + l[0] + '**' , '\n\n',
          const.replace('%%', l[1].split('/')[-1].split('?')[0].replace('-mini','')),
          '\n\n', '**Refactoring Guru:** '+ l[2], '\n\n', '**Meu Resumo: **')
    print()
```
