# Transitos pdf formatter

Após receber um pdf com os transito do meu mapa natal, deve retornar os dados codificados. Apesar de gerar muta coisa, é em ingles e é desorganizado o pdf

## Schema do passo a passo

+ 1 - Separar cada bloco de testo entre 'TRANSITING' e assim forma uma lista
+ 2 - Converter essa lista numa lista de objetos mapeando cada parte deses transitos
+ 3 - Duplicar para cada data, por exemplo,o primeiro é `Exact 1 Apr 2022, Exact 7 Aug 2022, Exact 4 Jan 2023` entãovai dividir em 3
+ 3 - Ordenar essa lista por tempo (dia, mes, ano, dia da semana)
+ 4 - Montar um Readme

## Funciona

+ Dá muito bug coms as datas, pois pode pegar mais de uma linha que o normal
+ Não está perfitio mas funciou corretamente par ao mes de novembro que eu estava querendo

```python
def get_transiction_title_and_house(text):
    idx_start = text.find('(')
    idx_end = text.find(')')
    
    title = 'TRANSITING ' + text[:idx_start].strip()
    house = text[idx_start:idx_end].replace('\n', '')
    
    return title, house

def day_strings(text):
    idx_end = text.find(')')
    text = text[idx_end:].split('\n')[1].strip()
    return text

def day_strings_alt(text):
    idx_end = text.find(')')
    text_re = text[idx_end:].split('\n')[1].strip()
    text_rest = text[idx_end:].split('\n')[2]
    text_rest_2 = text[idx_end:].split('\n')[3]
    
    # pois as datas podem ser grandes, entoa tem que pegar mais linhas 
#     print(text_rest)
    if('20' in text_rest):
        print('entrou 1')
        text_re += ' ' + text_rest
        
        
    # pois as datas podem ser grandes, entoa tem que pegar mais linhas 
#     print(text_rest_2)
    if('20' in text_rest_2):
        print('entrou 2')
        text_re += ' ' + text_rest_2
        
    return text_re.replace('2 0', '20')

def text_strings(text):
    idx_end = text.find(')')
    text = " ".join(text[idx_end:].split('\n')[2:]).replace('  ',' ')
    return text

def print_like_json(adict):
    print(json.dumps(adict,indent=2))
    
    
mapping_date_errors = {
    ' Exact 29 Feb': ' Exact 29 Feb 2024',
    ' Exact 14 Feb': 'Exact 14 Feb 2024',
    ' Exact 24': 'Exact 24 Dec 2023',
    ' Exact 4 Jan': 'Exact 4 Jan 2024',
}
    
def format_brief_date(exact_dt_str):
    print('exact_dt_str ORIGINAL:', repr(exact_dt_str))
    
    if(exact_dt_str in mapping_date_errors.keys()):
        print('TROCA POIS VAI DAR ERROR')
        exact_dt_str = mapping_date_errors[exact_dt_str]

    astr = exact_dt_str.strip().replace('  ',' ').replace('Exact ','').replace('Exac t','').strip()
    splited_date = astr.split(' ')
    convert_month = ''
    
    try:
        convert_month = adict_date[astr.split(' ')[1]]
    except Exception as e:
        print('exact_dt_str:', repr(exact_dt_str))
        print('astr:', astr)
        print('splited_date', splited_date)
        print(e)
        # substitui pois pegou a data errada
        
#         if(exact_dt_str == ):
#             print('ENTROU')
#             exact_dt_str = 
#         if(exact_dt_str == ):
#             print('ENTROU')
#             exact_dt_str = 
#         if(exact_dt_str == ):
#             print('ENTROU')
#             exact_dt_str = 
        
            
#         astr = exact_dt_str.strip().replace('  ',' ').replace('Exact ','').replace('Exac t','').strip()
#         splited_date = astr.split(' ')
#         convert_month = adict_date[astr.split(' ')[1]]
            
#         quit()
    
    try:
        python_date = datetime.datetime(
            int(splited_date[2]),
            int(convert_month),
            int(splited_date[0])
        )
    except Exception as e:
        print('ERRO AO CRIAR TIME DATE')
        print('TEXTO DE ENTRADA:', exact_dt_str)
        print('convert_month:', convert_month)
        print('splited_date', splited_date)
        print(e)
        
#         if(exact_dt_str == ' Exact 14 Feb'):
#             print('ENTROU')
#             exact_dt_str = 
#         if(exact_dt_str == ' Exact 24'):
#             print('ENTROU')
#             exact_dt_str = 'Exact 24 Dec 2023'
#         if(exact_dt_str == ' Exact 4 Jan'):
#             print('ENTROU')
#             exact_dt_str = 'Exact 4 Jan 2024'
            
#         # REFAZ COM NOVO DALOR
#         astr = exact_dt_str.strip().replace('  ',' ').replace('Exact ','').replace('Exac t','').strip()
#         splited_date = astr.split(' ')
#         convert_month = adict_date[astr.split(' ')[1]]
#         python_date = datetime.datetime(
#             int(splited_date[2]),
#             int(convert_month),
#             int(splited_date[0])
#         )
#         quit()
    
    week_day = day_week_dict[str(python_date.strftime("%a"))]
    
    return splited_date[2] + \
        '-' + adict_date[splited_date[1]] + \
        '-' + splited_date[0].zfill(2) + \
        ' - ' + week_day

all_months = []

def struct_transiction(text):
    global count
    count = count + 1
    alist = []
    adict = {}
    title, house = get_transiction_title_and_house(text)
    adict['title'] = title
    adict['house'] = house + ')'
    adict['text'] = text_strings(text)
    dates = list(map(format_brief_date,
                     day_strings_alt(text).split(',') 
    ))
    for adate in dates:
        copy_dict = copy.deepcopy(adict)
        copy_dict['date'] = adate
        mes = "-".join(adate.split('-')[:2])
        copy_dict['month'] = mes
        all_months.append(mes)
        alist.append(copy_dict)
    return alist
```

```python
def day_strings_alt(text):
    try:
        
        idx_end = text.find(')')
        text_re = text[idx_end:].split('\n')[1].strip()
        
        if( len(text[idx_end:].split('\n')) >= 3 ):
            text_rest = text[idx_end:].split('\n')[2]
            if('20' in text_rest or '2 0' in text_rest and not '20s' in text_rest):
                print('entrou 1')
                text_re += ' ' + text_rest
        
        if( len(text[idx_end:].split('\n')) >= 4 ):
            text_rest_2 = text[idx_end:].split('\n')[3]
            if('20' in text_rest_2 or '2 0' in text_rest_2 and not '20s' in text_rest_2):
                print('entrou 2')
                text_re += ' ' + text_rest_2
            
    except Exception as e:
        print('ERROR', text)
        print('erro no alt:', e)
        
    return text_re.replace('2 0', '20').replace('1 9', '19').replace('Exa ct', 'Exact')
```

```python
month_now = ultimate_list[0]['month']

with open("README-GLOBAL.md",'w', encoding = 'utf-8') as f:
    f.write("# {}\n\n".format('Transitos 2022 a 2023'))
    for el in ultimate_list:
        # Imprimir meses
        if(el['month'] != month_now):
            month_now = el['month']
            f.write("## {}\n\n".format(month_now))
        # Imprimie Postagem
        # data
        f.write("### {}\n\n".format(el['date']))
        # body
        f.write("**TITLE**: {}\n\n".format(el['title']))
        f.write("**HOUSE**: {}\n\n".format(el['house']))
        f.write("{}\n\n".format(el['text']))
```
